```python
### Remove extreme outliers using IQR
cols = [ "DistanceFromHome",'Age', 'MonthlyIncome']

for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
##Does overtime increase attrition?
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="OverTime", hue="Attrition")
plt.show()

pd.crosstab(df["OverTime"], df["Attrition"], normalize="index")*100

###Employees who work overtime are more likely to leave the company than those who do not.
```

```python
###Does monthly income affect attrition?

plt.figure(figsize=(7,5))
sns.boxplot(data=df,
            x="Attrition",
            y="MonthlyIncome",showfliers=False)
plt.show()

##Employees who left the company generally have lower monthly incomes than those who stayed. 
# This suggests that lower salary may be associated with a higher likelihood of employee attrition, although salary alone is not the only factor influencing turnover.
```

```python
###Which department has the highest attrition?

plot_df = df.copy()

plot_df["Department"] = "Human Resources"

plot_df.loc[
    plot_df["Department_Research and Development"] == 1,
    "Department"
] = "Research and Development"

plot_df.loc[
    plot_df["Department_Sales"] == 1,
    "Department"
] = "Sales"

sns.countplot(
    data=plot_df,
    x="Department",
    hue="Attrition"
)

plt.xticks(rotation=20)
plt.show()

pd.crosstab(
    plot_df["Department"],
    plot_df["Attrition"],
    normalize="index"
) * 100

##Although the Research and Development department has the highest number of employees leaving the company, it also has the largest workforce. When attrition is analyzed as a percentage, 
# Human Resources has the highest attrition rate (23.08%), followed by Sales (21.58%), 
# while Research and Development has the lowest attrition rate (14.83%). This indicates that employees in the Human Resources and Sales departments are more likely to leave the company.
```

```python
Does Job Satisfaction affect Attrition?
sns.countplot(data=df,
              x="JobSatisfaction",
              hue="Attrition")
plt.show()
```

```python
Does Job Satisfaction affect Attrition?
pd.crosstab(
    df["JobSatisfaction"],
    df["Attrition"],
    normalize="index"
) * 100


##There is a clear relationship between job satisfaction and employee attrition. Employees with the lowest job satisfaction (Level 1) have the highest attrition rate (24.62%), 
# while employees with the highest job satisfaction (Level 4) have the lowest attrition rate (12.44%). This suggests that improving employee job satisfaction may help reduce employee turnover.
```

```python
###Does Age affect Attrition?
bins = [18, 25, 35, 45, 60]
labels = ["18-25", "26-35", "36-45", "46-60"]

df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)

pd.crosstab(
    df["AgeGroup"],
    df["Attrition"],
    normalize="index"
) * 100


###Employee attrition is highest among younger employees. The 18–25 age group has the highest attrition rate (35.45%), while employees aged 36–45 have the lowest attrition rate (10.40%). 
# This indicates that younger employees are more likely to leave the company, whereas mid-career employees tend to be more stable and remain with the organization.
```

```python
####Does Business Travel affect Attrition?

## Travel:70 left / 260 total = ~27% attrition rate.
##Rarely:150 left / 950 total = ~15% attrition rate.

plot_df = df.copy()

plot_df["BusinessTravel"] = "Non-Travel"

plot_df.loc[
    plot_df["BusinessTravel_Travel_Rarely"] == 1,
    "BusinessTravel"
] = "Travel_Rarely"

plot_df.loc[
    plot_df["BusinessTravel_Travel_Frequently"] == 1,
    "BusinessTravel"
] = "Travel_Frequently"

sns.countplot(
    data=plot_df,
    x="BusinessTravel",
    hue="Attrition"
)

plt.xticks(rotation=25)
plt.show()

pd.crosstab(
    plot_df["BusinessTravel"],
    plot_df["Attrition"],
    normalize="index"
) * 100

##Business travel appears to have a significant relationship with employee attrition. Employees who travel frequently have the highest attrition rate (26.54%), 
#while employees who do not travel have the lowest attrition rate (8.70%). This suggests that frequent business travel may increase the likelihood of employees leaving the company.
```

```python
###Does Job Level affect Attrition?
sns.countplot(data=df,
              x="JobLevel",
              hue="Attrition")
plt.show()

pd.crosstab(
    df["JobLevel"],
    df["Attrition"],
    normalize="index"
) * 100
```

```python
###Does TrainingTimesLastYear affect Attrition?
sns.countplot(data=df,
              x="TrainingTimesLastYear",
              hue="Attrition")
plt.show()

"""Employee attrition is highest among employees at Job Level 1 (26.43%), indicating that entry-level employees are the most likely to leave the company. 
Attrition decreases substantially for higher job levels,
with Job Level 2 showing the lowest attrition rate (9.49%). 
Although Job Level 5 has a relatively high attrition rate (20%), this result should be interpreted cautiously due to the small number of employees at that level."""
```

```python
sns.barplot(
    data=df,
    x="TrainingTimesLastYear",
    y="Attrition"
)
plt.show()


"""Employees who received no training during the last year had the highest attrition rate (approximately 28%).
In general, employees who received more training tended to have lower attrition rates,
suggesting that training and employee development may contribute to improving employee retention.
However, the relationship is not perfectly linear, and some categories may be influenced by smaller sample sizes."""
```

```python
plot_df = df.copy()

plot_df["MaritalStatus"] = "Divorced"

plot_df.loc[
    plot_df["MaritalStatus_Married"] == 1,
    "MaritalStatus"
] = "Married"

plot_df.loc[
    plot_df["MaritalStatus_Single"] == 1,
    "MaritalStatus"
] = "Single"

sns.countplot(
    data=plot_df,
    x="MaritalStatus",
    hue="Attrition"
)

plt.show()

pd.crosstab(
    plot_df["MaritalStatus"],
    plot_df["Attrition"],
    normalize="index"
) * 100

"""Marital status appears to be associated with employee attrition. Single employees have the highest attrition rate (26.48%), which is nearly twice the rate of Married (13.55%) and 
Divorced (11.15%) employees. 
This suggests that single employees are more likely to leave the company than employees with other marital statuses."""
```
